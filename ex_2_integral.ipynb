```python
##Exercise 2

import random
import math

xmin = 0
xmax = 2
ymin = 1
ymax = 9


def f(x):
    return x ** 3 + 1


def estimate_integral(n):
    m = 0
    for _ in range(n):
        xp = (xmax-xmin) * random.random() + xmin
        yp = (ymax-ymin) * random.random() + ymin

        if yp < f(xp):
            m += 1
    sampling_area = (xmax - xmin) * (ymax - ymin)  # This is 2 * 8 = 16

    integral_estimate = (m / n) * sampling_area

    return integral_estimate


def relative_error(integral_estimate):
    return ((abs(integral_estimate - 6)) / 6) * 100  # Calculate positive relative error

# List of n values
n_values = [10**4, 10**5, 10**6, 10**7]

# Lists to store results
int_estimates = []
errors = []

# Loop through the n_values, estimate Pi, and print the results
for n in n_values:
    integral_estimate = estimate_integral(n)
    rel_error = relative_error(integral_estimate)
    
    int_estimates.append(integral_estimate)
    errors.append(rel_error)

    print(f"Estimated value of integral for n={n} is {integral_estimate}")
    print(f"Relative error for n={n} is {rel_error}%\n")
```

```python
#Graphical View

import matplotlib.pyplot as plt


# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, errors, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Experiments (n)')
plt.ylabel('Relative Error (%)')
plt.title('Relative Error vs. Number of Experiments')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()


```

```python
##Exercise 2

# List of n values
n_values = [10**4, 10**5, 10**6, 10**7, 10**8]

# Function to run experiments and store results
def run_experiment(series):
    # Loop through the n_values, estimate Pi, and print the results
    for n in n_values:
        integral_estimate = estimate_integral(n)
        rel_error = relative_error(integral_estimate)

        int_estimates.append(integral_estimate)
        errors.append(rel_error)

        print(f"Estimated value of integral for n={n} is {integral_estimate}")
        print(f"Relative error for n={n} is {rel_error}%\n")


# Initialize lists to store results for each series
Series1 = []
Series2 = []
Series3 = []
Series4 = []
Series5 = []


run_experiment(Series1)
run_experiment(Series2)
run_experiment(Series3)
run_experiment(Series4)
run_experiment(Series5)

# Output the results
print("Series1:", Series1)
print("Series2:", Series2)
print("Series3:", Series3)
print("Series4:", Series4)
print("Series5:", Series5)
```
